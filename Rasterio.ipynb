```python
import rasterio

# 1. Abrir o arquivo (isso cria uma conexão com o arquivo no disco)
with rasterio.open('/home/jovyan/Downloads/paisagem.tiff') as src:
    
    # 2. Ler as propriedades geográficas (Metadados)
    print("Quantidade de Bandas:", src.count)
    print("Largura (pixels):", src.width)
    print("Altura (pixels):", src.height)
    print("Sistema de Coordenadas (CRS):", src.crs)
    print("Transformação Geoespacial (Coordenadas dos cantos):\n", src.transform)
    
    # 3. Ler os pixels reais da imagem de uma banda específica (ex: Banda 1)
    # Isso transforma os pixels em uma matriz numérica (NumPy Array)
    dados_pixels = src.read(1)

# Fora do bloco 'with', o arquivo foi fechado com segurança, 
# mas a variável 'dados_pixels' continua disponível para você usar!
print("\nTipo do dado na memória:", type(dados_pixels))
print("Formato da matriz de pixels:", dados_pixels.shape)
```
